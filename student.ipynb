```python
# The schema as I understand it from the above exploratory prodding and some online research...
# id: just a unique identifier for a house/purchase AFAICT
# date: purchase date in month/day/year format note this isn't necessarily mm/dd/yyyy but stuff like 6/29/2011 also
# price: price in US dollars
# bedrooms: number of bedrooms, integer
# bathrooms: number of bathrooms, floating point number, full bathrooms count for 1, half bathrooms count for .5, and powder rooms for .25
# sqft_living: square feet of living space
# sqft_lot: square footage of overall property/lot
# floors: number of floors the house has, floating point, wtf are half floors?
# waterfront: 1 if house has waterfront view, 0 otherwise, NaN if missing
# view: an integer from 0 to 4 rating the view from the house, NaN if missing, presumably bigger numbers mean a better view
# condition: integer from 1 to 5 describing the houses condition, presumably bigger is better
# grade: overall grade given to the house, an integer rating from 3 to 13 inclusive, presumably bigger means better
# sqft_above: square footage of the house not counting basement
# sqft_basement: square footage of basement area
# yr_built: year house was built
# yr_renovated: year house was renovated, may be NaN
# zip code: self explanatory
# lat: latitude of house
# long: longitude of house
# sqft_living15: average living area of 15 closest properties
# sqft_lot_15: average square footage of overall property/lot area
```
